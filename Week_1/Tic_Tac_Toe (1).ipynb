```python
def output_of_game(game):
  for i in game:
    print(" | ".join(i))
    print("----------")
```

```python
def is_gamespace_filled(game):
    for i in game:
        if " " in i:
            return False
    return True
```

```python
def game_winner(game, player_moves):
    check= [ # row wise check

        [game[0][0], game[0][1], game[0][2]],
        [game[1][0], game[1][1], game[1][2]],
        [game[2][0], game[2][1], game[2][2]],

             #column wise check

        [game[0][0], game[1][0], game[2][0]],
        [game[0][1], game[1][1], game[2][1]],
        [game[0][2], game[1][2], game[2][2]],
             #diagonal check

        [game[0][0], game[1][1], game[2][2]],
        [game[2][0], game[1][1], game[0][2]]
             ]

    return [player_moves, player_moves, player_moves] in check


# Algorithm

def minimax(game, max_checking):
    if game_winner(game, 'O'):
        return 1

    elif game_winner(game, 'X'):
        return -1

    elif is_gamespace_filled(game):
        return 0

    if max_checking:
        best_score = float('-inf')

        for i in range(3):
            for j in range(3):

                if game[i][j] == " ":
                    game[i][j] = 'O'
                    score = minimax(game, False) # simulate player's turn

                    game[i][j] = " "
                    best_score = max(score, best_score)
        return best_score

    else: # Player's turn
        best_score = float('inf')

        for i in range(3):
            for j in range(3):

                if game[i][j] == " ":
                    game[i][j] = 'X'
                    score = minimax(game, True) #simulate computers response
                    game[i][j] = " "
                    best_score = min(score, best_score)

        return best_score




def best_move(game):

    best_score = float('-inf')
    move = None

    for i in range(3):
        for j in range(3):

            if game[i][j] == " ":
                game[i][j] = 'O'
                score = minimax(game, False) #simulate player
                game[i][j] = " "

                if score > best_score:
                    best_score = score
                    move_value = (i, j)
    return move_value


def main_game():

    game = [[" ", " ", " "],
            [" ", " ", " "],
            [" ", " ", " "]
            ]
    output_of_game(game)

    while True:
        # User Side:
        while True:

            print("Rows and columns are numbered as 0, 1 and 2.")
            row = int(input("Choose Row: "))
            col = int(input("Choose Column: "))

            if game[row][col] == " ":
                game[row][col] = 'X'
                break

            else:
                print("Wrong move, Try again.")

        output_of_game(game)

        if game_winner(game, 'X'):
            print("You WIN!")
            break

        elif is_gamespace_filled(game):
            print("A Tie! Better luck next time!")
            break

        # Computer Side:

        position = best_move(game)
        game[position[0]][position[1]] = 'O'

        print("Computer's Turn:")
        output_of_game(game)

        if game_winner(game, 'O'):
            print("Computer Won!")
            break

        elif is_gamespace_filled(game):
            print("A Tie! Better luck next time!")
            break

main_game()
```
